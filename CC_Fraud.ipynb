## Fraud Detection Patterns
Non fraudulent transaction have a significantly different average value compared to the non fraudulent, however fraudulent transactions' amounts vary more (their amounts have higher interquartile range). 

*In order to make the patterns in the data more interpretable Amounts have been log transformed.*

```python
# log transform Amount and box-plot
cc_fraud['Amount_log'] = np.log(cc_fraud.Amount + 0.01)
sns.boxplot(x="Class", y="Amount_log", data=cc_fraud)
plt.show()
```

```python
cc_fraud.groupby('Class').Amount_log.std()
```

```python
# we can reject the null hypothesis that fraudulent and non fraudulent amounts have equal variances
stats.levene(cc_fraud.loc[cc_fraud.Class==0, 'Amount'], cc_fraud.loc[cc_fraud.Class==1, 'Amount'])
```

```python
# the average amount is different across classes
# given that we are dealing with fairly big sample sizes a Welchâ€™s t-test with non qual variances should give non biased results
stats.ttest_ind(cc_fraud.loc[cc_fraud.Class==0, 'Amount'], cc_fraud.loc[cc_fraud.Class==1, 'Amount'], equal_var=False)
```

```python
# plot the distributions of the top few variables that have the highest absolute difference in either mean, median
#  or standard deviation between the non fraudulent and the fraudulent class
eda = cc_fraud.drop(['Amount', 'Amount_log', 'Time'], axis=1)\
    .groupby('Class')\
    .aggregate(['mean', 'median', 'std'])\
    .diff()\
    .T\
    .reset_index()
eda['abs_diff'] = eda[eda.columns[3]].abs()
eda_sub = eda.sort_values('abs_diff', ascending=False).head(10)
iter_vars = eda_sub.level_0.unique()
```

```python
import matplotlib.gridspec as gridspec
# iter_vars=['V17','V14', 'V11', 'V4', 'V15', 'V13']
features = iter_vars
nplots=np.size(features)
plt.figure(figsize=(15,4*nplots))
gs = gridspec.GridSpec(nplots,1)
for i, feat in enumerate(features):
    ax = plt.subplot(gs[i])
    sns.distplot(cc_fraud[feat][cc_fraud.Class==1], bins=30)
    sns.distplot(cc_fraud[feat][cc_fraud.Class==0],bins=30)
    ax.legend(['fraudulent', 'non-fraudulent'],loc='best')
    ax.set_xlabel('')
    ax.set_title('Distribution of feature: ' + feat)
```
